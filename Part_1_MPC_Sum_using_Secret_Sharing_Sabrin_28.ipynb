```python
import random

# Number of parties and their private values
num_parties = 3
private_values = [100, 70, 50]  # Data of Ahmed, Ali, Rashida

# Function to split secret into random shares
def split_secret(value, n):
    shares = [random.randint(0, 100) for _ in range(n-1)]
    final_share = value - sum(shares)
    shares.append(final_share)
    return shares

# Distributing shares for each party
shares = [split_secret(val, num_parties) for val in private_values]

# Transpose shares for proper distribution
distributed_shares = list(zip(*shares))

# Each party sums the shares they received
partial_sums = [sum(parts) for parts in distributed_shares]

# Calculate the final total sum
final_sum = sum(partial_sums)

print("Private values:", private_values)
print("Shares distributed:", shares)
print("Partial sums by party:", partial_sums)
print("Final sum (should be total of private values):", final_sum)
```
