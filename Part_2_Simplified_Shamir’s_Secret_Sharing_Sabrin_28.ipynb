```python
from sympy import symbols, Eq, solve
import random

# Setting values and threshold k and number of participants n
secret = 1234
k = 3  # Minimum shares for reconstruction
n = 5  # Number of participants

# Create a random polynomial
coeffs = [secret] + [random.randint(1, 100) for _ in range(k-1)]

def f(x):
    return sum([coeffs[i] * (x ** i) for i in range(k)])

# Generate shares (points)
shares = [(i, f(i)) for i in range(1, n+1)]

print("Shares:", shares)

# Reconstruct secret using k shares (e.g., first 3)
x = symbols('x')
equations = [Eq(sum([symbols('a{}'.format(j)) * (sh[0]**j) for j in range(k)]), sh[1]) for sh in shares[:k]]
coeff_solutions = solve(equations)

reconstructed_secret = coeff_solutions[symbols('a0')]
print("Reconstructed secret:", reconstructed_secret)
```
